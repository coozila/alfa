```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import base64
from io import BytesIO
```

```python
height = [10, 20, 3, 32, 23, 15, 32, 111]
y = [1, 2, 3, 4, 5, 6, 7, 8]

cmap = mpl.cm.get_cmap('Set2_r')
vmin, vmax = min(height), max(height)
norm = mpl.colors.Normalize(vmin = vmin, vmax = vmax)

colors = []
for i in height:
    val = norm(i)
    colors.append(cmap(val))

fig, (ax) = plt.subplots(figsize = (8, 5), ncols = 1)
ax.bar(y, height, color = colors, width = 0.5)

ax.set_title("Predicted Cluster Classes", fontsize = 20)
plt.ylabel('Class Count', fontsize = 16)
plt.xlabel('Class', fontsize = 16)
fig.tight_layout()

buf = BytesIO()
fig.savefig(buf, format="png")
data = base64.b64encode(buf.getbuffer()).decode("utf-8")
```
