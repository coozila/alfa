```python
import matplotlib.pyplot as plt
import base64
from io import BytesIO
```

```python
fig, ax = plt.subplots()
ax.axis('tight')
ax.axis('off')

cols = ['Metric', 'Value']
rows = [['MAE', 1], ['MSE', 2], ['RMSE', 3], ['R**2', 4], ['Adj. R**2', 5]]

the_table = ax.table(cellText=rows, colWidths=[0.1, 0.2 ],
                     colLabels=cols, loc = 'center')
the_table.set_fontsize(16)
the_table.scale(3, 3)

buf = BytesIO()
fig.savefig(buf, format="png")
data = base64.b64encode(buf.getbuffer()).decode("ascii")
```
